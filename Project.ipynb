# Visualization

# Number of Releases per Year on Netflix

```python
release_year_counts = data['release_year'].value_counts().sort_index()

mean_count = release_year_counts.mean()

plt.figure(figsize=(10, 8))
plt.barh(release_year_counts.index, release_year_counts.values, color='blue')

plt.axvline(x=mean_count, color='black', linestyle='--', label=f'Mean Count: {mean_count:.2f}')

plt.title('Number of Releases per Year on Netflix')
plt.xlabel('Number of Releases')
plt.ylabel('Release Year')

plt.legend()

plt.show()
```

# TV SHOW VS MOVIES On NETFLIX

```python
plt.figure(figsize=(4, 4))
sns.countplot(data=data, x='type')
plt.title('TV Shows vs Movies on Netflix')
plt.show()
```

# Type of shows percentage on Netflix

```python
labels=["Movie","TV Show"]
size=data["type"].value_counts()
colors = ['#E50914', '#000000']
explode=[0,0.1]
plt.figure(figsize=(4, 4))
plt.rcParams["figure.figsize"]=(9,9)
plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True,textprops=dict(color="white", weight="bold", fontsize=11), startangle=90)
plt.title("Movies vs. TV Shows on Netflix")
plt.legend()
plt.show()
```

# Netflix Rating Distribution

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='rating', order=data['rating'].value_counts().index)
plt.title('Netflix Rating Distribution')
plt.xticks(rotation=45)
plt.show()
```

# Top 5 countries with the highest Movies/TV Shows

```python
top_countries = data['country'].value_counts().head(5)
plt.figure(figsize=(4, 4))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Countries with Highest Movies/TV Shows')
plt.show()
```

# Top 5 Netflix country distribution separated by type

```python
top_5_countries = data[data['country'].isin(top_countries.index)]
plt.figure(figsize=(4, 4))
sns.countplot(data=top_5_countries, x='country', hue='type')
plt.title('Top 5 Netflix Countries by Type of Release (Movies vs TV Shows)')
plt.show()
```

#  Top 10 genres of movies and tv shows

```python
genres = data['genre'].str.split(',').explode().str.strip()
top_genres = genres.value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=top_genres.index, x=top_genres.values)
plt.title('Top 10 Genres on Netflix')
plt.show()
```

# Best time to release a movie/TV show

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')

data['month'] = data['date_added'].dt.month
data['year'] = data['date_added'].dt.year

heatmap_data = data.pivot_table(index='year', columns='month', aggfunc='size', fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='d')
plt.title('Best Time to Release a Movie/TV Show (Heatmap)')
plt.show()
```
